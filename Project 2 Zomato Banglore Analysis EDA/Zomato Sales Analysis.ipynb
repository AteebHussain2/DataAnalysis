```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
df = pd.read_csv("zomato.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop(['url', 'phone', 'address', 'dish_liked', 'reviews_list', 'menu_item'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
df.info()
```

# Dropping Duplicates

```python
df.drop_duplicates(inplace = True)
df.shape
```

# Cleaning Rate Column

```python
df['rate'].unique()
```

### Removing New, /5 and - in rate column

```python
def handlerate(value):
    if(value == 'NEW' or value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

# Handling Null Values

```python
df.isnull().sum()
```

### Filling Rate Null Values

```python
df['rate'].mean()
```

```python
df['rate'] = df['rate'].fillna(df['rate'].mean())
df['rate'].isnull().sum()
```

```python
df.info()
```

### Dropping Null Values

```python
df.dropna(inplace = True)
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)' : 'cost2plate', 'listed_in(type)' : 'Type'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

#### Dropping Column 'listed_in(city)' and using 'location' instead because they both are same

```python
df = df.drop(['listed_in(city)'], axis = 1)
```

```python
df.head()
```

```python
df['cost2plate'].unique()
```

```python
def handlecomma(value):
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)

df['cost2plate'] = df['cost2plate'].apply(handlecomma)
df['cost2plate'].unique()
```

### Cleaning Location Folder

```python
df['location'].value_counts()
```

```python
locations = df['location'].value_counts(ascending = False)
locations
```

```python
locations_lessthan300 = locations[locations < 300]
locations_lessthan300
```

```python
def handlelocation(value):
    if value in locations_lessthan300:
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handlelocation)
df['location'].value_counts()
```

### Cleaning Rest Type Column

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types < 1000]
rest_types_lessthan1000
```

```python
def handleresttype(value):
    if value in rest_types_lessthan1000:
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handleresttype)
df['rest_type'].value_counts()
```

### Cleaning Cuisines Column

```python
df['cuisines'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending = False)
cuisines
```

```python
cuisines_lessthan100 = cuisines[cuisines < 100]
cuisines_lessthan100
```

```python
def handlecuisines(value):
    if value in cuisines_lessthan100:
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handlecuisines)
df['cuisines'].value_counts()
```

```python
df['cuisines'].head()
```
